```python
# Imports
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

%matplotlib inline
```

```python

#genshin impact @gamewith.net
urls = {
    "all_character_list":"https://gamewith.net/genshin-impact/article/show/22357"
}

# Use requests to retrieve data from a given URL
response = requests.get(urls['all_character_list'])

# Parse the whole HTML page using BeautifulSoup
response_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Title of the parsed page
response_soup.title
```

```python
# We can also get it without the HTML tags
response_soup.title.string
```

```python
# Selector

# select div containing table of character list
div_characters = response_soup.find('div', {'class': 'genshin_cha'})
```

```python
# print readable result
print(div_characters.prettify())
```

```python
# access html tag
table_characters = div_characters.table
print(table_characters.prettify())
```

```python
for  row in table_characters.find_all('tr', {'class': 'w-idb-element'}):
  print(row)
  # access html attribut
  print(row['data-col1'])
  print(row['data-col2'])
  print(row['data-col3'])
  print(row['data-col4'])
  print(row['data-col5'])
  # method 1
  img_character = row.td.img['data-original']
  # method 2
  img_character_2 = row.find_all('img', {'class':'w-article-img'})[0]['data-original'] #in this case, the row contains 2 images, 0 index is the first image
  print(img_character)
  print(img_character_2)
```

```python
for  head in table_characters.find_all('th'):
  print(head)
  print(head.name)
  print(head.attrs)
  # get it without the HTML tags
  print(head.string)
  print(head.text)
  
```

```python

def ignoreImg(tag):
    print(tag)
    return not (tag in ['Anemo','Pyro','Geo','Hydro','Cryo','Electro'] )

for  row in table_characters.find_all('tr', {'class': 'w-idb-element'}):
  img_character_3 = row.find_all('img', {'class':'w-article-img','width':'50px','height':'50px'})
  img_character_4 = row.find_all('img', alt=ignoreImg)
  print(img_character_4)


  
```
